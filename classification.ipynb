```python
import pandas as pd
import numpy as np
import re
```

```python
df_raw = pd.read_csv("csv/csv_cleaned.csv")
df_raw.head(10)
```

```python
df = df_raw

def url_to_sentence(url):
    url = re.sub(r'^https?://', '', url)         # Remove http:// or https://
    url = re.sub(r'[^a-zA-Z0-9]', ' ', url)       # Replace symbols with space
    url = re.sub(r'\s+', ' ', url)                # Replace multiple spaces with single space
    return url.strip().lower()                    # Lowercase and trim

df['url_sentence'] = df['website_url'].astype(str).apply(url_to_sentence)

df['cleaned_website_text_cleaned'] = df['cleaned_website_text'].str.lower().apply(lambda x: re.sub(r'[^a-z\s]', '', x))
df.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stop_words = list(stopwords.words('english'))

vectorizer = CountVectorizer(stop_words=stop_words, max_df=0.95, min_df=5)
X = vectorizer.fit_transform(df['url_sentence'])

# Get the list of keywords
keywords = vectorizer.get_feature_names_out()

# Get the frequencies
keyword_freq = X.sum(axis=0).A1
freq_df = pd.DataFrame({'keyword': keywords, 'frequency': keyword_freq})
freq_df = freq_df.sort_values(by='frequency', ascending=False)
```
